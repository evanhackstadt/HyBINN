```python
# Plot losses recorded from specified runs in experiments/
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Run 4

```python
run_dir = "../experiments/runs/binn_20260316_lr1e3_wd1e4/"
csv_path = os.path.join(run_dir, f"run4.csv")

df = pd.read_csv(csv_path, index_col=0)
df.head()
```

```python
# convert to long format for seaborn plotting
dfm = df.melt(id_vars='epoch', var_name='type', value_name='loss')
dfm.head()
```

```python
ax = sns.lineplot(dfm, x='epoch', y='loss', hue='type')
ax.set_title("Run 4")

fig = ax.get_figure()
fig.savefig(os.path.join(run_dir, f"run4_loss.png"))
```

```python
ax = sns.lineplot(df, x='epoch', y='cindexes')
ax.set_title("Run 4")

fig = ax.get_figure()
fig.savefig(os.path.join(run_dir, f"run4_cindex_zoom.png"))
```

```python
ax = sns.lineplot(df, x='epoch', y='cindexes')
ax.set_ylim(0, 1)
ax.set_yticks([0, 0.25, 0.5, 0.75, 1.0])
ax.set_title("Run 4")

fig = ax.get_figure()
fig.savefig(os.path.join(run_dir, f"run4_cindex.png"))
```
